```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

# NC 다이노스 선수 데이터 분석
## 1. 데이터 적재 및 통합

```python
data = pd.read_excel('data/NC Dinos.xlsx',sheet_name=None)
NC13, NC14, NC15 = data.values()
```

```python
NC13['연도'] = 2013
NC14['연도'] = 2014
NC15['연도'] = 2015
```

```python
NC13.drop('Unnamed: 0',axis=1,inplace = True)
NC14.drop('Unnamed: 0',axis=1,inplace = True)
NC15.drop('Unnamed: 0',axis=1,inplace = True)
```

```python
NCAll = pd.concat([NC13,NC14,NC15],axis=0,ignore_index=True)
```

```python
NCAll.head()
```

## 2. 2015년도 선수 기록을 '타율'로 내림차순 정렬

```python
NC15.sort_values(by='타율',ascending=False)
```

```python
# 타석 수가 200 이상인 타자를 대상으로 타율 순으로 내림차순
NC15[NC15.타석 >= 200].sort_values(by=['타율','안타'],ascending=[False,False])
```

## 3.  (실습) 2013~2015년  선수별로 3년 간 평균 안타수를 구한 후, 평균 안타수가 가장 많은 상위 10명의 선수를 출력

```python
top10 = NCAll.pivot_table(index='선수명',
                          aggfunc='mean',
                          values='안타').sort_values(by='안타',
                                                    ascending=False).head(10)
top10
```

```python
# 선수별로 실제 뛴 시즌의 수와 3년 동안 친 전체 안타수 추가
# 사실 계층 색인이라 이걸 굳이 계층 색인으로 처리할 필요없어서 단층으로
data = NCAll.pivot_table(index='선수명',aggfunc=['mean','count','sum'],values='안타')
data.columns = ['평균안타수','참여시즌수','전체안타수']
data.sort_values(by='평균안타수',ascending=False).head()
```

## 4. (실습) 3에서 구한 상위 10명의 타자들의 3년간 안타, 홈런, 타율을 출력

```python
top10_3year = NCAll[['선수명','연도','안타','홈런','타율']]
top10_3year.set_index(['선수명','연도'],inplace=True)
top10_3year = top10_3year.unstack('연도')
top10_3year.loc[top10.index].fillna('-').sort_values(by='선수명')
```

## 5. (실습) 년도별 팀의 기록 (총 안타수, 총 홈런수, 총 도루수)을 출력

```python
nc_total = NCAll.pivot_table(index='연도',aggfunc='sum',values=['안타','홈런','도루'])
nc_total
```
